```python
# Select only 1 Star Review

df_1star = df_review[df_review["rating"] == "1 star"]
df_1star_prompt =df_1star.iloc[0:5,]
```

```python
# Send Unknown list to GPT

%pip install openai
%pip install python-dotenv
import openai
import os
# Init the .env file
from dotenv import load_dotenv
load_dotenv()
import json

with open("/content/OPENAI_CRED_ISB.json") as config_file:
  config = json.load(config_file)
  USERNAME = config['USERNAME']
  APIKEY = config['APIKEY']
  ORGANIZATION = config['ORGANIZATION']

openai.organization = ORGANIZATION
openai.api_key = APIKEY

## check permission of using models
#openai.Model.list()
```

```python
df_1star_prompt
```

```python

# Batch size for each API request
batch_size = 1

# Calculate the total number of batches required
num_batches = (len(df_1star_prompt) + batch_size - 1) // batch_size

# Create an empty DataFrame to store the final output
df_output = pd.DataFrame() ### Open this when it is first run . In case of runtime error , close this line to continue the work

# Loop through the data in batches and send API requests
for i in range(num_batches):
    start_index = i * batch_size
    end_index = (i + 1) * batch_size
    batch_data = df_1star_prompt[start_index:end_index]

    # Create an empty list to store the product categories for the batch
    is_negative_positive_list = []
    category_list = []

    # Generate categories for each product in the batch
    for index, row in batch_data.iterrows():
        prompt = "Return 1 for negaive and 0 for positive and category of problem whether it is about cleanliness, food, customer services for this review ' " + f"\nReview: {row['review']}\nis_negaive: 'response here' \ncategory : 'respose here' "

        # Set the maximum tokens to limit the response length
        response = openai.Completion.create(
            engine='text-davinci-003',
            prompt=prompt,
            max_tokens=50,
            stop=None,
            temperature=0.7,
        )

  # Get the generated response choice
        choice = response.choices[0]
      # Split the choice text into lines

        choice_text = choice.text.strip()

     # Split the choice text by comma
        choice_parts = choice_text.split(',')

    if len(choice_parts) >= 3:
      is_negative_positive = choice_parts[0].strip()
      category = choice_parts[1].strip()
    else:
      print("Response format not as expected:", choice_text)  # Debugging line

    # Append the product sentiment and category to the lists
      is_negative_positive_list.append(is_negative_positive)
      category_list.append(category)

    # Add the 'Category' column to the batch DataFrame
      batch_data['is_negative'] = is_negative_positive
      batch_data["category"] = category

    # Concatenate the batch DataFrame to the final output DataFrame
      df_output = pd.concat([df_output, batch_data])

    # Process the API response (e.g., print response status code)
    #print(f"Batch {i + 1} - Status Code: {response.status_code}")
      print(len(df_output))
# Print the DataFrame with the new 'Category' column
```
